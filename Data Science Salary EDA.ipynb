# What is the average employee salary?

```python
df['salary_in_usd'].mean()
```

# Top 5 countries where the employees are residing?

```python
df['employee_residence'].value_counts().head()
```

# What is the Job title of the highest paid person?

```python
df[df['salary_in_usd'] == df['salary_in_usd'].max()]
```

# What is the job title of the lowest paid person? 

```python
df[df['salary_in_usd'] == df['salary_in_usd'].min()]
```

# Highest count of jobs and their respective salaries.

```python
salary_title_df = df.groupby("job_title").agg(
    mean_salary=('salary_in_usd', 'mean'), count_jobs=('job_title', 'count')
).reset_index()
salary_title_df.sort_values(by="count_jobs", ascending=False).head(10)
```

# What was the average BasePay of all Job titles per year? (2020-2022) ?

```python
df.groupby('work_year').salary_in_usd.mean()
```
